```python
from ast import literal_eval
```

```python
def eval_(x):
    #print(x)
    if (type(x) != float) and (len(x)>1):
        r = literal_eval(x)
    else:
        r = x
    return r
```

```python
df_api['media'] = df_api.media.apply(lambda x: eval_(x))
```

```python
import urllib.request
```

```python
def get_pic(platformId, c, url):
    path = None
    try: 
        path = 'fb_pics_total/'+platformId + '_'+ str(c) +'.jpg'
        urllib.request.urlretrieve(url, path)
    except:
        pass
    return path
```

```python
def get_url_pic(medialist,platformId):
    #print(medialist)
    
    
    if (type(medialist) != float) and (len(medialist)>1):
        try:
            path_list = []
            medialist = literal_eval(medialist)
            c = 0
            for m in medialist:
                c += 1
                #print(m)
                if m['type'] == 'photo':
                    p = get_pic(platformId, c, m['url'])
                    path_list.append(p)
        except:
            path_list = []
    else:path_list = []
        
    if len(path_list) == 0:
        path_list = None
        
    else:
        pass
    return path_list
```

```python
df_api['path'] = df_api.apply(lambda x: get_url_pic(x.media, x.platformId),axis =1 )
```

```python
def count_media(d_list):
    v_c = 0
    p_c = 0
    other_c = 0
    if (type(d_list) != float) and (len(d_list)>1):
        d_list = literal_eval(d_list)
        for m in d_list:
            if m['type'] == 'video':
                v_c += 1
            elif m['type'] == 'photo':
                p_c += 1
            else:
                other_c += 1
    return v_c, p_c, other_c
```

```python
df_api['video_count'] = df_api.media.apply(lambda x: count_media(x)[0])
df_api['photo_count'] = df_api.media.apply(lambda x: count_media(x)[1])
df_api['othermedia_count'] = df_api.media.apply(lambda x: count_media(x)[2])
```
